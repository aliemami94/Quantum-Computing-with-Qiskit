```python
import numpy as np
from qiskit import  transpile, QuantumCircuit
from qiskit_aer import Aer

```

```python
qc = QuantumCircuit(2,2)
qc.h(0) #apply the Hadamard to the q0 
qc.cx(0,1) #apply the CNOT to the target q1 and contorol q0  
qc.measure([1,0],[1,0])
qc.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
```

```python
backend = Aer.get_backend('qasm_simulator')
tqc = transpile(qc, backend)
job = backend.run(tqc,shots=1000)
result = job.result()
counter = result.get_counts(qc)
 
print(counter)
```

```python
import numpy as np
from qiskit import  transpile, QuantumCircuit
from qiskit_aer import Aer
```

```python
"add the hadamard to make the uniform state"
qc = QuantumCircuit(5,5)
qc.h(1)
qc.h(2)
qc.x(0)
qc.h(0)
qc.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
```

```python
qc.h(0)
qc.cx(1,0)
qc.tdg(0)
qc.cx(2,0)
qc.t(0)
qc.cx(1,0)
qc.tdg(0)
qc.cx(2,0)
qc.t(0)
qc.tdg(1)
qc.h(0)
qc.cx(2,1)
qc.tdg(1)
qc.cx(2,1)
qc.s(1)
qc.t(2)

qc.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
```

```python
qc.h(1)
qc.h(2)
qc.x(1)
qc.x(2)
qc.h(1)
qc.cx(2,1)
qc.h(1)
qc.x(2)
qc.h(2)
qc.x(1)
qc.h(1)
qc.measure([2,1],[2,1])
qc.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
```

```python
backend = Aer.get_backend('qasm_simulator')
tqc = transpile(qc, backend)
job = backend.run(tqc,shots=100)
result = job.result()
counter = result.get_counts(qc)
 
print(counter)
```
